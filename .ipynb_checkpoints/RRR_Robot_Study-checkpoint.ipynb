## Find Jacobian (Skew theory)

```python
T1 = Rz(theta1)*Tz(L1)
R1 = T1[:3, :3]
O1 = T1[:3, -1]
```

```python
T2 = Ry(theta2)*Tx(L2)
T12 = T1 * T2
R2 = T12[:3, :3]
O2 = T12[:3, -1]
```

```python
T3 = Ry(theta3)*Tx(L3)
T13 = T1 * T2 * T3
R3 = T13[:3, :3]
O3 = T13[:3, -1]
```

```python
H = T1 * T2 * T3
R = H[:3, :3]
O = H[:3, -1]
```

```python
O0 = Matrix([0,0,0])
z1 = Matrix([0,0,1])
z1
```

```python
z2 = T1
z2 = z2[:3,1]
z3 = T1*T2
z3 = z3[:3,1] #first column because rotation is around Y axis
z2
```

```python
z3
```

```python
# The first joint (revolute)
Jv1 = z1.cross(O - O0);
Jw1 = z1;
```

```python
# The second joint (revolute)
Jv2 = z2.cross(O - O1);
Jw2 = z2;
Jv2
```

```python
# The third joint (revolute)
Jv3 = z3.cross(O - O2);
Jw3 = z3;
Jv3
```

```python
Jg = simplify(Matrix([[Jv1, Jv2, Jv3],
                      [Jw1, Jw2, Jw3]]))

Jg
```

## Find Trajectory
**Joint trajectory q(t) from q(0) = (0, 0, 0) to q(2) = (2, 3, 4) with null initial and final velocities and accelerations. (polynomial)**

```python
t0 = 0.
tf = 2.

q0 = np.array([0, 0, 0])
qf = np.array([2, 3, 4])

v0 = np.array([0, 0, 0])
vf = np.array([0, 0, 0])
```

```python
i=0
[q0[i], v0[i], qf[i], vf[i]]
```

```python
A = np.array([[1, t0, t0**2, t0**3],
              [0, 1, 2*t0, 3*t0**2],
              [1, tf, tf**2, tf**3],
              [0, 1, 2*tf, 3*tf**2]])
a0, a1, a2, a3 = [], [], [], []
for i in range(3):
    c = np.array([q0[i], v0[i], qf[i], vf[i]])
    b = np.linalg.inv(A).dot(c)
    a0.append(b[0])
    a1.append(b[1])    
    a2.append(b[2])    
    a3.append(b[3])    
```

```python
t = np.arange(t0, tf, 0.01)

fig1, axes1 = plt.subplots(nrows=3,ncols=3, figsize=(15, 15))
fig1.subplots_adjust(hspace=.7,wspace=0.8)

for i in range(3):
    
    q = a0[i] + a1[i]*t + a2[i]*t**2 + a3[i]*t**3
    v = a1[i] + 2*a2[i]*t + 3*a3[i]*t**2
    acc = 2*a2[i] + 6*a3[i]*t
    
    axes1[i,0].set_title('Position vs. Time for Joint'+str(i))
    axes1[i,0].plot(t, q)
    axes1[i,1].set_title('Speed vs. Time for Joint'+str(i))
    axes1[i,1].plot(t, v)
    axes1[i,2].set_title('Acceleration vs. Time for Joint'+str(i))
    axes1[i,2].plot(t, acc)

# plt.savefig("question2.png")
plt.show()
```

## Trapezoidal Point-to-Point

**Joint trajectory for the following commands: $PTP$  $q1 = (0, 0, 0)$ to $q2 = (2, 3, 4)$
(trapezoidal)**

**o Controller command interpretation frequency  $f = 100 Hz$**

**o Maximum joint velocity  $1 rad/s$**

**o Maximum joint acceleration  $10 rad/s^2$**

```python
delta_t = 0.01
n=0
while np.floor(delta_t*10**n) != delta_t*10**n:
    n+=1
E = 10**-n

n
```

```python
q0 = np.array([0, 0, 0])
qf = np.array([2, 3, 4])

v0 = np.array([0, 0, 0])
v = np.array([1, 1, 1])

acc0 = np.array([0, 0, 0])
acc = np.array([10, 10, 10])
```

```python
ta = v/acc
ta
```

```python
ta_new = np.array([])
tf_new = np.array([])

for i in range(len(ta)):
    ta_new = np.append(ta_new, np.round(ta[i], n))
    if np.remainder(ta[i], delta_t) != 0:
        ta_new[i] += E
tf = (qf-q0)/v + ta_new
for i in range(len(tf)):     
    tf_new = np.append(tf_new, np.round(tf[i], n))
    if np.remainder(tf[i], delta_t) != 0:
        tf_new[i] += E
    
```

```python
ta_new_max = max(ta_new)
tf_new_max = max(tf_new)

tf_new_max
```

```python
v_new = (qf-q0)/(tf_new_max-ta_new_max)

acc_new = v_new/ta_new_max
```

```python
print(v_new)
print(acc_new)
```

```python
# joints coeffiecents
# t0 --> ta
b00 = q0
b01 = v0
b02 = 0.5*acc_new

# ta --> tf-ta
b10 = q0 + 0.5*acc_new*ta_new_max**2 - v_new*ta_new_max
b11 = v_new

# tf-ta --> tf
b20 = qf - 0.5*acc_new*tf_new_max**2
b21 = acc_new*tf_new_max
b22 = -0.5*acc_new
```

```python
t = np.arange(t0, tf_new_max, delta_t)

q_equations = np.empty((0,len(t)), np.float)
v_equations = np.empty((0,len(t)), np.float)
acc_equations = np.empty((0,len(t)), np.float)

for i in range(3):
    
    q_equations = np.append(q_equations, [(b00[i] + b01[i]*t + b02[i]*t**2)*(t<=ta_new_max)\
                        +(b10[i]+b11[i]*t)*(t>ta_new_max)*(t<=(tf_new_max-ta_new_max))\
                        +(b20[i]+b21[i]*t+b22[i]*t**2)*(t>(tf_new_max-ta_new_max))*(t<=tf_new_max)], axis=0)
    v_equations = np.append(v_equations, [(b01[i]+2*b02[i]*t)*(t<=ta_new_max)\
                                         +b11[i]*(t>ta_new_max)*(t<=(tf_new_max-ta_new_max))\
                                         +(b21[i]+2*b22[i]*t)*(t>(tf_new_max-ta_new_max))*(t<=tf_new_max)], axis=0)
    acc_equations = np.append(acc_equations, [2*b02[i]*(t<=ta_new_max)\
                                             +0*(t>ta_new_max)*(t<=(tf_new_max-ta_new_max))\
                                             +2*b22[i]*(t>(tf_new_max-ta_new_max))*(t<=tf_new_max)], axis=0)
    
```

```python
t = np.arange(t0, tf_new_max, delta_t)

fig1, axes1 = plt.subplots(nrows=3,ncols=3, figsize=(15, 15))
fig1.subplots_adjust(hspace=.7,wspace=0.8)

for i in range(3):
    
    q = q_equations[i]
    v = v_equations[i]
    acc = acc_equations[i]
    
    axes1[i,0].set_title('Position vs. Time for Joint'+str(i))
    axes1[i,0].plot(t, q)
    axes1[i,1].set_title('Speed vs. Time for Joint'+str(i))
    axes1[i,1].plot(t, v)
    axes1[i,2].set_title('Acceleration vs. Time for Joint'+str(i))
    axes1[i,2].plot(t, acc)

plt.savefig("question3.png")
plt.show()
```

## Trapezoidal - Straight Line Movement

**Joint trajectory for the following commands: $LIN$  $p1 = (1, 0, 1)$ to $p2 = (\frac{\sqrt{2}}{2}, \frac{\sqrt{2}}{2}, 1.2)$
(trapezoidal)**


**o Controller command interpretation frequency  $f = 100 Hz$**

**o Maximum linear velocity  $1 m/s$**

**o Maximum linear acceleration  $10 m/s^2$**

```python
def RRR_IK(pnt, L=[1,1,1]):
    x = pnt[0]
    y = pnt[1]
    z = pnt[2]
    L1 = 1.
    L2 = 1.
    L3 = 1.
    
    theta1 = np.arctan2(y, x)
    X = np.sqrt(x**2 + y**2)
    Y = (z - L1)
    theta3 = np.arccos((X**2 + Y**2 - L2**2 - L3**2) / (2.*L2*L3))

    theta2 = - np.arctan2(L2 + L3*np.cos(theta3), L3*np.sin(theta3))\
             - np.arctan2(-Y, np.sqrt(x**2 + y**2))
    return [theta1, theta2, theta3]
```

```python
def calculate_jacobian(thetas):
    L1 = 1
    L2 = 1
    L3 = 1
    theta1, theta2, theta3 = thetas
    T = Rz(theta1)*Tz(L1)*Ry(theta2)*Tx(L2)*Ry(theta3)*Tx(L3)
    # diff by theta1
    Td1 = Rzd(theta1)*Tz(L1)*Ry(theta2)*Tx(L2)*Ry(theta3)*Tx(L3) * rotation_to_transformation_matrix(R.T)
    # extract 3 components for the 1st joint
    J1r = Matrix([Td1[0,3], Td1[1,3], Td1[2,3]])
    # diff by theta2
    Td2 = Rz(theta1)*Tz(L1)*Ryd(theta2)*Tx(L2)*Ry(theta3)*Tx(L3) * rotation_to_transformation_matrix(R.T)
    # extract 3 components for the 2nd joint
    J2r = Matrix([Td2[0,3], Td2[1,3], Td2[2,3]])
    # diff by d3
    Td3 = Rz(theta1)*Tz(L1)*Ry(theta2)*Tx(L2)*Ryd(theta3)*Tx(L3) * rotation_to_transformation_matrix(R.T)
    # extract 3 components for the 3rd joint
    J3r = Matrix([Td3[0,3], Td3[1,3], Td3[2,3]])

    # Full Jacobian 6x3
    Jn = Matrix([[J1r, J2r, J3r]])
    return np.array(Jn).astype(np.float)
```

```python
delta_t = 0.01
tf = 10
time = np.arange(0, 100)#tf, delta_t)

p0 = np.array([1,0,1])
pf = np.array([1./np.sqrt(2), 1./np.sqrt(2), 1.2])


# positions
x = ((pf[0] - p0[0])/tf) * time + p0[0]
y = ((pf[1] - p0[1])/tf) * time + p0[1]
z = ((pf[2] - p0[2])/tf) * time + p0[2]
```

```python
# velocities
# at t0 and tf: vel=0 for all joints, at waypoints velocities are constants:
vx = ((pf[0] - p0[0])/tf)
vy = ((pf[1] - p0[1])/tf)
vz = ((pf[2] - p0[2])/tf)

vel = np.array([vx, vy, vz])
```

```python
joint_config = np.zeros((3,len(time))) 
joint_vel = np.zeros((3,len(time)))

for i in range(len(time)): 
    points = [x[i], y[i], z[i]] 
    joint_config[:, i] = RRR_IK(points) 
    if i == 1 or i == len(time)-1: 
        joint_vel[:, i] = 0 
    else:
        J = calculate_jacobian(joint_config[:, i]); 
         
        joint_vel[:,i] = np.linalg.inv(J).dot(vel.T)
```
